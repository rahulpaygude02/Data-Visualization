```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.express as px
from plotly.offline import plot
import plotly as py
import plotly.tools as tls
import cufflinks as cf

from  plotly.offline import iplot
cf.go_offline()
```

#line Plots

```python
import plotly.graph_objects as go

px.line(df1, x='A',y='B',labels={'x':'A','y':'B'})
```

```python
df_stocks = px.data.stocks()
df_stocks
```

```python
px.line(df_stocks, x='date',y='GOOG',labels={'x':'date','y':'Price'})
```

```python
px.line(df_stocks, x='date',y=['GOOG','AAPL'],labels={'x':'date','y':'Price'},title = 'Apple vs Google')
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df_stocks.date , y = df_stocks.AAPL, mode = 'lines' , name = 'Apple'))

fig.add_trace(go.Scatter(x=df_stocks.date , y = df_stocks.AMZN, mode = 'lines+markers' , name = 'Amazon'))
```

# scatter plot


```python
df_iris = px.data.iris()
px.scatter(df_iris,x='sepal_width',y='sepal_length',color='species',size='petal_length',hover_data=['petal_width'])
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=df_iris.sepal_width,y=df_iris.sepal_length,mode ='markers',marker_color=df_iris.sepal_width,
        text=df_iris.species,marker=dict(showscale=True)))
```

# heatmaps

```python
flights = sns.load_dataset('flights')
flights

```

```python
fig = px.density_heatmap(flights,x='year',y='month',z='passengers',
                         color_continuous_scale = 'Viridis')
fig
```

# 3d scatter plot

```python
fig = px.scatter_3d(flights,x='year',y='month',z='passengers',color='year'
                    ,opacity=0.9)
fig
```

```python
fig = px.line_3d(flights,x='year',y='month',z='passengers',color='year'
                    )
fig
```

# Map scatter plot

```python
df=px.data.gapminder().query('year==2007')
fig = px.scatter_geo(df,locations= 'iso_alpha',
                    color = 'continent',hover_name='country',
                    size='pop',projection = 'orthographic')
fig
```

# polar chart

```python
df_wind = px.data.wind()
px.scatter_polar(df_wind,r='frequency',theta ='direction',color='strength',
             size='frequency',symbol='strength',template = 'plotly_dark')
```

```python
df_wind
```

```python
df_wind = px.data.wind()
px.line_polar(df_wind,r='frequency',theta ='direction',color='strength',
             line_close=True,template = 'plotly_dark')
```

# Animated plot

```python
df_cnt =px.data.gapminder()
px.scatter(df_cnt,x='gdpPercap',y = 'lifeExp',size='pop',color ='continent',
          animation_frame ='year',animation_group='country',
           hover_name = 'country',log_x=True,range_x=[100,100000],
          range_y=[25,90])
```
