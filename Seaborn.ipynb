```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set(style = 'darkgrid')
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.relplot(x='total_bill',y='tip',data = tips)
```

```python
sns.relplot(x='total_bill',y='tip',data = tips,hue='smoker',style='time')
```

```python
sns.relplot(x='total_bill',y='tip',data = tips,hue='smoker',style='time',size='size')
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri)
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', 
            data = fmri, ci = 'sd')
```

```python
sns.relplot(x = 'timepoint', y = 'signal', estimator = None,
            kind = 'line', data = fmri)
```

```python
sns.relplot(x = 'timepoint', y = 'signal', hue = 'event',
            kind = 'line', data = fmri)
```

```python
sns.relplot(x = 'timepoint', y = 'signal', hue = 'region', 
            style = 'event', kind = 'line', data = fmri)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', 
            col = 'time', data = tips)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker',
            col = 'size', data = tips)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', col = 'size',
            data = tips, col_wrap=3, height=3)
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
sns.catplot(x = 'color', y = 'price', kind = 'boxen', 
            data = diamonds.sort_values('color'))
```

```python
sns.catplot(x = 'total_bill', y = 'day', hue = 'sex', 
            kind = 'violin', data = tips, split = True,)
```

```python
g = sns.catplot(x = 'day', y = 'total_bill', kind = 'violin', 
                inner = None, data = tips)
sns.swarmplot(x = 'day', y = 'total_bill', color = 'k', 
              size = 3, data = tips, ax = g.ax)
```

```python
def sinplot():
    x = np.linspace(0, 14, 100)
    for i in range(1,5):
        plt.plot(x, np.sin(x))
```
